```python
# MAT 421 Spring 2022
# Nick Florez
# Module H 22.1-22.3

#Problem 1
import numpy as np
#import scipy for solving initial value problem
from scipy.integrate import solve_ivp
#import matplotlib for plotting curves
import matplotlib.pyplot as plt
from functools import partial
plt.style.use("seaborn-poster")

#function
def my_logistic_eq(t, P, r, K):
    dP = r * P * (1 - P / K)
    return dP

#Test cases
dP = my_logistic_eq(0, 10, 1.1, 15)
print(dP)

#Plotting
t0 = 0
tf = 20
P0 = 10
r = 1.1
K = 20
t = np.linspace(0, 20, 2001)
f = partial(my_logistic_eq, r = r, K = K)
sol = solve_ivp(f, [t0,tf], [P0], t_eval = t)
plt.figure(figsize = (10, 8))
plt.plot(sol.t, sol.y[0])
plt.plot(t, K * P0 * np.exp(r * t)/(K + P0 *(np.exp(r * t)-1)),"r:")
plt.xlabel("time")
plt.ylabel("population")
plt.legend(["Numerical Solution", "Exact Solution"])
plt.grid(True)
plt.show()
```

```python
#Problem 2
import numpy as np

def my_lorenz(t, S, sigma, rho, beta):
    if (sigma < 0) or (rho < 0) or (beta < 0):
        raise BaseException("Error: Sigma, rho, beta needs to be all non-negative")
    dS = np.zeros(3)
    dS[0] = sigma * (S[2] - S[1])
    dS[1] = S[0] * (rho - S[2]) - S[1]
    dS[2] = S[0] * S[1] - beta * S[2]
    return dS

s = np.array([1, 2, 3])
dS = my_lorenz(0, s, 10, 28, 8/3)
print(dS)
```
