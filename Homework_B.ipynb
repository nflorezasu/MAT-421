```python
#MAT 421 Fall 2022
#Homework B - Nick Florez

#Problem 7: 
#import functions for using to graph plots
import matplotlib.pyplot as plt
import numpy as np
from scipy import interpolate

def my_interp_plotter(x,y,X,option):
    #First we set the graph to have certain parameters
    plt.figure(figsize = (10,8))
    plt.plot(x,y,'ro',label = "data")
    f = interpolate.interp1d(x,y,option)
    Y=f(X)
    
    #Next we give the graph labels for x, y, and the "option" chosen
    plt.plot(X,Y,"b",label ="option")
    plt.xlabel = ("X")
    plt.ylabel = ("Y")
    plt.legend()

    if option == "nearest":
        plt.title("Nearest interpolation of data")
        #If option is nearest, then the graph should print the nearest interpolation
    elif option == "linear":
        plt.title("Linear interpolation of data")
        #If option is linear, then the graph should print the linear interpolation
    else:
        plt.title("Cubic interpolation of data")
        #If option is cubic, then the graph should print the cubic spline interpolation
        plt.show()
```

```python
#Test Cases:

x = np.array([0,0.1,0.15,0.35,0.6,0.7,0.95,1])
y = np.array([1,0.8187,0.7408,0.4966,0.3012,0.2466,0.1496,0.1353])

my_interp_plotter(x,y,np.linspace(0,1,101),"nearest")
my_interp_plotter(x, y, np.linspace(0, 1, 101), "linear")
my_interp_plotter(x, y, np.linspace(0, 1, 101), "cubic")
```

```python
#Problem 8

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

def my_D_cubic_spline(x,y,X,D):
    #Set the CubicSpline graph so that it is actually "clamped"
    f = CubicSpline(x,y,bc_type = ((1,D),(1,D)))
    #new f(X) variable
    y_new = f(X)
    #plot new 
    plt.plot(X,y_new,"b",label="clamped")
    plt.scatter(x,y)
    plt.plot(1.5,f(1.5),'ro',ms=25)
    return y_new
```

```python
#Test cases
x = [0, 1, 2, 3, 4]
y = [0, 0, 1, 0, 0]
X = np.linspace(0, 4, 101)

# Solution: Y = 0.54017857
Y = my_D_cubic_spline(x, y, 1.5, 1) 

plt.figure(figsize = (10, 8))
plt.subplot(221)
plt.plot(x, y, "ro", X, my_D_cubic_spline(x, y, X, 0), "b")
plt.subplot(222)
plt.plot(x, y, "ro", X, my_D_cubic_spline(x, y, X, 1), "b")
plt.subplot(223)
plt.plot(x, y, "ro", X, my_D_cubic_spline(x, y, X, -1), "b")
plt.subplot(224)
plt.plot(x, y, "ro", X, my_D_cubic_spline(x, y, X, 4), "b")
plt.tight_layout()
plt.show()
```
